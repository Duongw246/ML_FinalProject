```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
```

```python
data = pd.read_csv('E:\Learning\Machine learning\Finalterm\data\diabetes_012_health_indicators_BRFSS2021.csv')
data.head()
```

```python
X = data.drop(['Diabetes_012'], axis=1)
y = data['Diabetes_012']
```

```python
ros = RandomOverSampler(random_state=0)
X_res, y_res = ros.fit_resample(X, y)
print(Counter(y_res).items())
X_res.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.24)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score

# we can add class_weight='balanced' to add panalize mistake
model = KNeighborsClassifier()

model.fit(X_train, y_train)

predict = model.predict(X_test)# check performance
print('Accuracy score:',accuracy_score(y_test, predict))
print('F1 score:',f1_score(y_test, predict, average='weighted'))
```
